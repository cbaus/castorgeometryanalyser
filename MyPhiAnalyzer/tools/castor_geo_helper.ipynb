```python
def plot(std, inner, new, title="Side View", stdlb="outer",innerlb="inner",newlb="new"):
    plt.plot([a[2] for a in std], [a[0] for a in std],color="black",label=stdlb)
    plt.plot([a[2] for a in std], [a[1] for a in std],color="black")

    if(inner):
        plt.plot([a[2] for a in inner], [a[0] for a in inner],color="blue",label=innerlb)
        plt.plot([a[2] for a in inner], [a[1] for a in inner],color="blue")

    if(new):
        plt.plot([a[2] for a in new], [a[0] for a in new],color="red",label=newlb)
        plt.plot([a[2] for a in new], [a[1] for a in new],color="red")

        print '<!-- ' + newlb + ' -->'
        for sec in new:
            print '<ZSection rMin="{0:.2f}*mm" rMax="{1:.2f}*mm" z="{2:.2f}*mm"/>'.format(sec[0],sec[1],sec[2])

    ax=plt.gca()
    ax.set_title(title)
    ax.set_xlabel("z [mm]")
    ax.set_ylabel("r [mm]")
    
    plt.legend(loc="upper left")
    
def slope (sections):
    rmin_slope=(sections[-1][0]-sections[1][0])/(sections[-1][2]-sections[1][2])
    rmax_slope=(sections[-2][1]-sections[0][1])/(sections[-2][2]-sections[0][2])
    return rmin_slope, rmax_slope
```
